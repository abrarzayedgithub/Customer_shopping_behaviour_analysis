```python
df['Review_Rating'] = df.groupby('Category')['Review_Rating'].transform(lambda x : x.fillna(x.median))
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
labels = ['young adult','adult','middle-aged','senior']
df['age_group'] = pd.qcut(df['age'],q=4, labels = labels)
```

```python
#create column purchase_frequency_days
frequency_mapping = {
                    'Weekly':7,
                    'Fortnightly':14,
                    'Monthly': 30,
                    'Querterly':90,
                    'Every 3 months':90,
                    'Bi-Weekly':14,
                    'Annually' : 365
}
```

```python
import pandas as pd
from sqlalchemy import create_engine

server = r'DESKTOP-CFA1C02\SQLEXPRESS'
database = 'customer_behaviour_analysis'
connection_url = f"mssql+pyodbc://@{server}/{database}?driver=ODBC+Driver+17+for+SQL+Server&trusted_connection=yes"
engine = create_engine(connection_url)

# Comprehensive search query for all table-like objects
diagnostic_query = """
SELECT schema_name(t.schema_id) AS SchemaName, t.name AS TableName
FROM sys.tables t
"""

try:
    diagnostic_df = pd.read_sql(diagnostic_query, engine)
    if diagnostic_df.empty:
        print("Confirmed: The database is completely empty. No tables exist yet.")
    else:
        print("Tables found with schemas:")
        print(diagnostic_df)
except Exception as e:
    print("Error scanning database:", e)

```
